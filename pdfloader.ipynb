```python
from langchain.document_loaders import PyPDFLoader
import time

# Function to extract and time PDF loading using PyPDFLoader
def extract_with_pypdf(pdf_path):
    loader = PyPDFLoader(pdf_path)
    
    start_time = time.time()
    documents = loader.load()  # Load PDF
    end_time = time.time()
    
    elapsed_time = end_time - start_time
    print(f"PyPDFLoader took {elapsed_time:.2f} seconds to load the PDF.")
    
    # Output the first page text as a sample
    print("First Page Content (PyPDFLoader):")
    print(documents[0].page_content)
    return documents

# Example Usage
pdf_path = '2019_001_annual_en.pdf'
extract_with_pypdf(pdf_path)
```

```python
from langchain.document_loaders import PDFPlumberLoader
import time

# Function to extract and time PDF loading using PDFPlumberLoader
def extract_with_pdfplumber(pdf_path):
    loader = PDFPlumberLoader(pdf_path)
    
    start_time = time.time()
    documents = loader.load()  # Load PDF
    end_time = time.time()
    
    elapsed_time = end_time - start_time
    print(f"PDFPlumberLoader took {elapsed_time:.2f} seconds to load the PDF.")
    
    # Output the first page text as a sample
    print("First Page Content (PDFPlumberLoader):")
    print(documents[0].page_content)
    return documents

# Example Usage
```

```python
pdfplumber_data = extract_with_pdfplumber(pdf_path)
```

```python
from langchain.document_loaders import PyMuPDFLoader
import time

# Function to extract and time PDF loading using PyMuPDFLoader
def extract_with_pymupdf(pdf_path):
    loader = PyMuPDFLoader(pdf_path)
    
    start_time = time.time()
    documents = loader.load()  # Load PDF
    end_time = time.time()
    
    elapsed_time = end_time - start_time
    print(f"PyMuPDFLoader took {elapsed_time:.2f} seconds to load the PDF.")
    
    # Output the first page text as a sample
    print("First Page Content (PyMuPDFLoader):")
    print(documents[0].page_content)
    return documents

# Example Usage
```

```python
pymupdf_data = extract_with_pymupdf(pdf_path)
```

```python
def compare_pdf_loaders(pdf_path):
    print("\n--- PyPDFLoader ---")
    extract_with_pypdf(pdf_path)

    print("\n--- PDFPlumberLoader ---")
    extract_with_pdfplumber(pdf_path)

    print("\n--- PyMuPDFLoader ---")
    extract_with_pymupdf(pdf_path)

# Call the comparison function with your PDF path
pdf_path = '2019_001_annual_en.pdf'
compare_pdf_loaders(pdf_path)
```
